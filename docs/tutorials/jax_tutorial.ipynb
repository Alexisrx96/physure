# Getting Started with MeasureKit and JAX

This tutorial demonstrates how to use `MeasureKit` with the JAX backend for high-performance, JIT-compiled physical simulations.

## Setup

First, ensure you have `measurekit` installed with the `jax` extra.

```python
import jax
import jax.numpy as jnp
from measurekit.domain.measurement.quantity import Quantity
from measurekit.core.dispatcher import BackendManager

# Ensure JAX backend is active or explicitly used
q = Quantity(jnp.array([1.0, 2.0, 3.0]), "m")
print(f"Quantity created with backend: {q.magnitude.__class__.__name__}")
print(q)
```

## JIT Compilation

MeasureKit quantities are registered as JAX Pytrees, meaning they can be passed directly into JIT-compiled functions.

```python
@jax.jit
def calculate_velocity(displacement, time):
    return displacement / time

disp = Quantity(jnp.array([100.0, 200.0]), "m")
time = Quantity(jnp.array([10.0, 20.0]), "s")

velocity = calculate_velocity(disp, time)
print("Velocities:", velocity)
```

## Automatic Vectorization (vmap)

You can also use `jax.vmap` with Quantities.

```python
def kinetic_energy(mass, velocity):
    return 0.5 * mass * velocity**2

# Vectorize over mass and velocity
vmapped_ke = jax.vmap(kinetic_energy)

masses = Quantity(jnp.ones(5), "kg")
velocities = Quantity(jnp.arange(1.0, 6.0), "m/s")

energies = vmapped_ke(masses, velocities)
print("Kinetic Energies:", energies)
```
